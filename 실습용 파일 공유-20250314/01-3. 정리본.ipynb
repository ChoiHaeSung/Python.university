```python
# 1. 필요한 라이브러리 불러오기
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier

# 2. 도미 (Bream) 데이터 정의
bream_length = [25.4, 26.3, 26.5, 27.3, 27.5,
                27.9, 28.0, 28.7, 30.0, 30.5,
                31.0, 31.5, 32.0, 32.0, 32.0,
                33.0, 33.0, 33.5, 33.5, 34.0,
                34.0, 34.5, 35.0, 35.0, 35.0,
                36.0, 36.0, 37.0, 38.5, 38.5,
                39.5, 41.0, 41.0, 42.0, 43.0]

bream_weight = [242.0, 290.0, 340.0, 363.0, 430.0,
                450.0, 500.0, 390.0, 450.0, 500.0,
                475.0, 500.0, 340.0, 600.0, 600.0,
                700.0, 700.0, 610.0, 650.0, 575.0,
                685.0, 620.0, 680.0, 700.0, 725.0,
                720.0, 714.0, 850.0, 925.0, 975.0,
                950.0, 955.0, 925.0, 975.0, 950.0]

# 3. 도미 산점도 시각화
plt.scatter(bream_length, bream_weight)
plt.xlabel('length')
plt.ylabel('weight')
plt.title('Relationship of bream length, weight')
plt.show()

# 4. 빙어 (Smelt) 데이터 정의
smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2,
                11.3, 11.8, 11.8, 12.0, 12.2,
                12.4, 13.0, 14.3, 15.0]

smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8,
                8.7, 10.0, 9.9, 9.8, 12.2,
                13.4, 12.2, 19.7, 19.9]

# 5. 도미 + 빙어 산점도 시각화
plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)
plt.xlabel('length')
plt.ylabel('weight')
plt.title('Relationship of bream and smelt\'s length, weight')
plt.show()

# 6. 두 생선 데이터 병합 (입력 데이터)
length = bream_length + smelt_length
weight = bream_weight + smelt_weight

# 2차원 배열로 만들기 (입력 데이터)
fish_data = [[l, w] for l, w in zip(length, weight)]

# 7. 정답 데이터 (도미=1, 빙어=0)
fish_target = [1] * 35 + [0] * 14
print("Target labels:", fish_target)

# 8. k-NN 모델 생성 및 훈련 (기본 k=5)
kn = KNeighborsClassifier()
kn.fit(fish_data, fish_target)

# 9. 훈련 데이터에 대한 정확도 출력
print("Training accuracy:", kn.score(fish_data, fish_target))

# 10. 길이 30cm, 무게 600g 생선 예측
prediction = kn.predict([[30, 600]])
print("Prediction for [30, 600]:", prediction)  # 예측 결과 (1 = 도미)

# 11. 예측 생선 포함 산점도 시각화
plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)
plt.scatter(30, 600, color='green', marker='^')  # 예측 생선
plt.xlabel('length')
plt.ylabel('weight')
plt.title('New fish (30, 600) prediction point')
plt.show()

# 12. 모델이 내부에 저장한 데이터 확인 (입력값과 타깃)
print("Model stored X (inputs):", kn._fit_X)
print("Model stored y (targets):", kn._y)

# 13. k 값을 49로 바꾼 모델 생성 및 훈련
kn49 = KNeighborsClassifier(n_neighbors=49)
kn49.fit(fish_data, fish_target)

# 14. 정확도 출력 (k=49로 할 경우 정확도는 도미 비율과 같아짐)
print("Training accuracy with k=49:", kn49.score(fish_data, fish_target))

# 15. 도미의 비율 출력 (35/49)
print("Ratio of bream (1) in dataset:", 35 / 49)
```
