### 지도 학습과 비지도 학습



* **지도 학습(Supervised Learning)** : 훈련 하기 위한 데이터와 정답을 모두 입력
  - 분류 : 이미지나 텍스트 등 입력 데이터를 미리 정의된 레이블(예: 고양이/개, 스팸/정상)로 분류
  - 회귀 : 연속적인 값을 예측하는 작업(예: 집값 예측, 주가 예측)



* **비지도 학습(Unsupervised Learning)** : 타깃 없이 입력 데이터만 사용하여 훈련
  - 정답을 사용하지 않으므로 무언가를 맞힐 수 없음
  - 데이터를 잘 파악하거나 변형하기 위해 사용
* **비지도 학습의 사례**
  - 군집화 : 데이터 내 자연스러운 그룹을 찾는 작업(예: 고객 세분화, 문서 클러스터링)
  - 차원 축소 : 데이터의 특성을 요약해 시각화하거나 분석하기 위한 방법(예: PCA, t-SNE)
  - 연관 규칙 학습 : 아이템 간의 관계를 분석해 구매 패턴 등을 찾아내는 작업(예: 장바구니 분석)

### 훈련 세트의 테스트 세트

* 도미와 빙어의 데이터와 타깃을 주고 훈련한 다음, 같은 데이터로 테스트 한다면 모두 맞히는 것은 당연
* 머신러닝 알고리즘의 성능을 제대로 평가하려면 **훈련 데이터와 평가에 사용할 데이터가 달라야 함**
---
* **테스트 세트(test set)** : 평가에 사용하는 데이터
* **훈련 세트(train set)** : 훈련에 사용되는 데이터


---
* 샘플을 훈련 세트 35개, 테스트 세트 14개로 분리

```python
# 도미와 빙어의 데이터를 합쳐서 하나의 리스트로 저장
# - http://bit.ly/bream_smelt
fish_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0, 
                31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0, 
                35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0, 9.8, 
                10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
fish_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0, 
                500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0, 
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0, 6.7, 
                7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]
```

```python
# 각 생선의 길이와 무게를 하나의 리스트로 담은 2차원 리스트 생성
fish_data = [[l,w] for l, w in zip(fish_length, fish_weight)]
fish_target= [1] * 35 + [0] *14
```

```python
print(fish_data)
```

```python
print(fish_target)
```

```python
from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier()
```

```python
# 생선 데이터에서 처음 35개와 나머지 14개를 선택하여 훈련 세트와 테스트 세트로 분리
train_input = fish_data[:35] # 훈련 세트 입력
train_target = fish_target[:35] # 훈련 세트 타깃

test_input = fish_data[35:]# 테스트 세트 입력
test_target = fish_target[35:] # 테스트 세트 타깃
```

```python
# 훈련 세트만으로 fit() 메서드를 호출하여 모델을 훈련
kn.fit(train_input, train_target)
# 테스트 세트로 score() 메서드를 호출하여 평가
kn.score(test_input, test_target)
```

### 샘플링 편항 (Sampling Bias)
* 훈련 세트와 테스트 세트에 샘플이 골고루 섞여 있지 않아 샘플링이 한쪽으로 치우치는 현상
---
* 잘못된 훈련 데이터
  - 마지막 14개를 테스트 세트로 떼어 놓으면 훈련 세트에는 빙어가 하나도 없음
* 대책 : 훈련 세트와 테스트 세트를 나눌 때 도미와 빙어가 골고루 섞이게 나눠야 함
