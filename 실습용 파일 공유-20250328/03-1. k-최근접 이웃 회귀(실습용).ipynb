###  지난 시간에



### 농어의 무게를 예측하라
* 농어를 마리당 가격으로 판매했을 때 무게에 대한 고객 항의 발생 ==> 마켓에서 농어를 무게 단위로 판매
* 공급처에서 보내온 농어의 무게 측정 오류 발생
  - 정상 농어 정보(길이, 높이, 두께)를 이용하여 잘 못 측정된 농어의 무게를 예측할 수 있을까?



### k-최근접 이웃 회귀

* **분류(Classification)** : 이미지나 텍스트 등 입력 데이터를 미리 정의된 레이블(예: 고양이/개, 스팸/정상)로 분류
* **회귀(Regression)** : 연속적인 값을 예측하는 작업(예: 집값 예측, 주가 예측)

### 데이터 준비
* [코드 복사]http://bit.ly/perch_data

```python
import numpy as np

# 56마리 농어의 길이와 무게 정보
perch_length = np.array([8.4, 13.7, 15.0, 16.2, 17.4, 18.0, 18.7, 19.0, 19.6, 20.0, 21.0,
       21.0, 21.0, 21.3, 22.0, 22.0, 22.0, 22.0, 22.0, 22.5, 22.5, 22.7,
       23.0, 23.5, 24.0, 24.0, 24.6, 25.0, 25.6, 26.5, 27.3, 27.5, 27.5,
       27.5, 28.0, 28.7, 30.0, 32.8, 34.5, 35.0, 36.5, 36.0, 37.0, 37.0,
       39.0, 39.0, 39.0, 40.0, 40.0, 40.0, 40.0, 42.0, 43.0, 43.0, 43.5,
       44.0])
perch_weight = np.array([5.9, 32.0, 40.0, 51.5, 70.0, 100.0, 78.0, 80.0, 85.0, 85.0, 110.0,
       115.0, 125.0, 130.0, 120.0, 120.0, 130.0, 135.0, 110.0, 130.0,
       150.0, 145.0, 150.0, 170.0, 225.0, 145.0, 188.0, 180.0, 197.0,
       218.0, 300.0, 260.0, 265.0, 250.0, 250.0, 300.0, 320.0, 514.0,
       556.0, 840.0, 685.0, 700.0, 700.0, 690.0, 900.0, 650.0, 820.0,
       850.0, 900.0, 1015.0, 820.0, 1100.0, 1000.0, 1100.0, 1000.0,
       1000.0])
```

* 산점도 그래프를 이용하여 데이터의 분포 확인

```python
import matplotlib.pyplot as plt
plt.scatter(perch_length, perch_weight)
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

* 제공받은 56마리 농어 데이터를 머신러닝 모델에 사용하기 위해 훈련 세트와 테스트 세트 생성

```python
from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = train_test_split(perch_length, perch_weight, random_state = 42)
```

```python
train_input.shape, test_input.shape
```

```python
test_input
```

* 모델에 입력되는 특성 데이터는 길이(length) 하나만 해당
  - 생성된 perch_length는 1차원 배열
  - 사이킷런의 모델 입력 데이터로 쓰기 위해서는 2차원 배열로 변환 필요 ==> numpy의 reshape() 메서드 사용

```python
train_input = train_input.reshape(42,1) # train_input.reshape(42, 1)
test_input = test_input.reshape(14,1) # train_input.reshape(14,1)
```

```python
train_input.shape, test_input.shape
```

### 결정 계수 ($ R^2$) : coefficient of determination
* 회귀 알고리즘에서 훈련 모델의 평가 점수
* 분류 알고리즘과 달리 정확한 값을 맞히는 것은 거의 불가능



* 타깃의 평균 정도를 예측하는 수준 : $ R^2$는 0에 가까워짐
* 예측이 타깃과 거의 비슷한 수준 : $ R^2$는 1에 가까워짐

* **KNeighborsRegressor** : k-최근접 이웃 회귀 알고리즘 구현 클래스

```python
from sklearn.neighbors import KNeighborsRegressor
knr = KNeighborsRegressor()
# k-최근접 이웃 회귀 모델 훈련
knr.fit(train_input, train_target)
```

```python
# 결정계수를 이용하여 예측한 정확도값 출력
knr.score(test_input, test_target)
```

* sklearn.metrics 패키지에서 여러 가지 측정 도구 제공
  - **mean_absolute_error** : 타깃과 예측의 절대값 오차를 평균하여 반환

```python
from sklearn.metrics import mean_absolute_error
test_prediction = knr.predict(test_input)
mae = mean_absolute_error(test_target, test_prediction)
```

```python
mae
```

### 과대적합 vs 과소적합
* **과대적합(overfitting)** : 훈련 세트의 점수가 좋은데 테스트 세트에서 점수가 굉장히 나쁘게 나오는 경우
  - 훈련 세트에만 잘 맞는 모델이기 때문에 새로운 샘플에 대한 예측에서 잘 동작하지 않을 가능성이 높음
* **과소적합(underfitting)** : 훈련 세트보다 테스트 세트의 점수가 높거나 두 점수가 모두 너무 낮은 경우
  - 모델이 너무 단순하여 훈련 세트에 적절히 훈련되지 않았음
  - 훈련 세트와 테스트 세트의 크기가 너무 작으면 발생할 수도 있음 (테스트 세트가 훈련 세트의 특징을 따르지 못함)
 
---
* 잘 만들어진 모델은 보통 **훈련 세트의 점수가 조금 더 높게 나와야 함**

```python
knr.score(train_input, train_target)
```

#### **[해결 방법]** : 모델을 조금 더 복잡하게 만들어 훈련 세트에 더 잘 맞는 모델 생성
  - k-최근접 이웃 회귀 알고리즘 : 이웃의 개수(k) 줄임
 


```python
# 이웃의 개수를 3으로 설정합니다
knr.n_neighbors = 3

# 모델을 다시 훈련합니다
knr.fit(train_input, train_target)
print(knr.score(train_input, train_target))
```

```python
# 테스트 세트의 점수 확인
# - 훈련 세트의 점수 98.04% 보다 테스트 세트의 점수는 97.46%로 약간 낮은 점수 기록
# ==> 과소적합 해결
print(knr.score(test_input, test_target))
```

### 확인문제

```python
# k-최근접 이웃 회귀 객체를 만듭니다
knr = KNeighborsRegressor()
# 5에서 45까지 x 좌표를 만듭니다
x = np.arange(5, 45).reshape(-1, 1)

# n = 1, 5, 10일 때 예측 결과를 그래프로 그립니다
for n in [1, 5, 10]:
    knr.n_neighbors = n
    knr.fit(train_input, train_target)
    prediction = knr.predict(x)
    plt.scatter(train_input, train_target)
    plt.plot(x, prediction)
    plt.title('n_neighbors = {}'.format(n))
    plt.xlabel('length')
    plt.ylabel('weight')
    plt.show()
```
