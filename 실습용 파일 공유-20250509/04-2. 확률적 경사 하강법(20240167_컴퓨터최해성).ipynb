* 당면한 문제 : **훈련 데이터가 한 번에 준비되는 것이 아니라 조금씩 전달됨**
* **해결방안 1**

* **해결방안 2**

  - (문제점) 데이터를 버릴 때 다른 데이터에 없는 중요한 데이터가 포함될 가능성이 있음
* **해결방안 3**
  - 기존의 훈련 모델을 버리지 않고 새로운 데이터에 대해서만 조금씩 훈련 **(점진적 학습/온라인 학습)**

#### [경사 하강법(Gradient Descent)]

* 아래의 그림에서 **J(w)가 최소가 되는 w**를 찾는 문제
* 이때 J(w)를 비용(Cost), 오차(Error)라고 함
* 오차의 종류
  - 평균 절대 오차(Mean Absolute Erorr, MAE), 평균 제곱 오차(Mean Squared Error, MSE), 평균 제곱근 오차(Root Mean Square Error, RMSE)
  - 예) 평균 제곱 오차(Mean Squared Error, MSE) : y = mx + b 일 때, J(m, b) = (1/n) * ∑(yᵢ - (mxᵢ + b))²



#### [확률적 경사 하강법(Stochastic Gradient Descent)]
* 경사를 따라 무작위하게(랜덤하게) 내려가는 방법
* 가장 가파른 경사를 따라 원하는 지점에 도달하는 것이 목표
* 가장 가파른 길을 찾아 내려오지만 조금씩 내려오는 것이 중요

#### [손실 함수(Loss Function)]
* 비용 함수(Cost Function)라고 부르기도 함
* 어떤 문제에서 머신러닝 알고리즘이 **얼마나 엉터리인지 측정**하는 기준
* 손실 함수의 값이 작을수록 좋은 성능
* 어떤 값이 최솟값인지 알 수 없기 때문에 가능한 많이 찾아보고 만족할만한 수준이면 STOP

##### **회귀에서의 손실**
* 예측과 실제 값의 차이
* 평균 절대 오차(Mean Absolute Erorr, MAE), 평균 제곱 오차(Mean Squared Error, MSE), 평균 제곱근 오차(Root Mean Square Error, RMSE) 등을 사용

##### **분류에서의 손실**
* 정답을 못맞춘 값을 이용
* **정확도**를 계산한 후 음수로 변경하여 손실 함수로 사용하면?
  - 샘플의 개수에 따라 손실 함수로 사용할 수 있는 값이 고정되는 단점이 있음
  - 샘플의 개수가 적다면 경사 하강법을 이용하여 조금씩 움직일 수가 없음
* 예) 4개의 샘플에서 도미와 빙어를 구분하는 이진 분류 문제 적용
  - 도미 : 양성 클래스(1), 빙어 : 음성 클래스(0)
  - 4개의 샘플에 대한 정확도 : 0, 0.25, 0.5, 0.75, 1
  - 손실값 : -1, -0.75, -0.5, -0.25, 0



#### [로지스틱 손실 함수(Logistic Loss Function]
* 로지스틱 회귀(Logistic Regression)에서 사용되는 이진 분류 문제의 손실 함수
* 이진 크로스엔트로피 손실 함수(Binary Cross-Entropy Loss Function)이라고도 함
* **예측값이 실제 값과 얼마나 차이나는지를 측정**하며, 값이 작을수록 모델이 더 잘 맞췄다는 의미

---
##### **4개의 샘플에서 도미와 빙어를 구분하는 이진 분류 문제 다시 적용**
* 4개 샘플의 예측 확률을 0.9, 0.3, 0.2, 0.8로 가정
* 각 샘플의 손실값 계산
  - 양성 클래스 손실값 : -(1 x 예측 확률)
  - 음성 클래스 손실값 : -(1 x (1 - 예측 확률))
* **예측 확률에 로그 함수를 적용**
  - 예측 확률의 범위(0 ~ 1)에 대응하는 로그 함수 값은 음수가 되어 최종 손실 값은 양수
  - 로그 함수는 0에 가까울수록 아주 큰 음수가 되기 때문에 손실을 아주 크게 만들어 모델에 큰 영향을 미칠 수 있음



##### **일반적인 로지스틱 손실 함수 수식 정의**
$$
L(y, \hat{y}) = -\left[y \cdot \log(\hat{y}) + (1 - y) \cdot \log(1 - \hat{y})\right]
$$
* y : 실제값 (0 또는 1)
* y^ : 예측 확률 (0 ~ 1 사이 값, sigmoid의 결과)

### SGDClassifier

* fish_csv_data 파일을 읽어와서 판다스 데이터프레임 생성

```python
import pandas as pd

fish = pd.read_csv('https://bit.ly/fish_csv_data')
```

* 입력 데이터(Species)와 타깃 데이터(Weight, Length, Diagonal, Height, Width) 분리

```python
fish_input = fish[['Weight','Length','Diagonal','Height','Width']].to_numpy()
fish_target = fish['Species'].to_numpy()
```

* 훈련 세트와 테스트 세트 분리

```python
from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = train_test_split(fish_input, fish_target, random_state=42)
```

* 훈련 세트와 테스트 세트의 특성을 표준화 전처리
  - 훈련 세트에서 학습한 통계 값으로 테스트 세트로 변환해야 함

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(train_input)
train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)
```

* SGDClassifier 클래스를 이용하여 확률적 경사 하강법 분류 모델 생성
  - **loss** : 손실 함수 종류('hinge'(SVM), 'log_loss'(로지스틱 회귀) 등), 기본값은 'hinge'
  - **penalty** : 규제 방식 ('l1', 'l2', 'elasticnet'), 기본값은 'l2'
  - **alpha** : 규제 강도, 값이 클수록 더 강한 규제, 기본값은 0.0001
  - **max_iter** : 최대 반복(epoch) 횟수, 기본값은 1000
  - **tol** : 조기 종료 허용 오차, 손실 개선이 이 값보다 작으면 학습을 중단, 기본값은 1e-3

```python
from sklearn.linear_model import SGDClassifier

sc =SGDClassifier(loss = 'log_loss', max_iter =10, random_state =42)
sc.fit(train_scaled, train_target)
```

* 훈련 세트와 테스트 세트의 점수 확인

```python
sc.score(train_scaled, train_target)
```

```python
sc.score(test_scaled, test_target)
```

* partial_fit() 메서드를 사용하여 **점진적(온라인) 학습** 적용
  - SGDClassifer 객체를 다시 만들지 않고 훈련한 모델에 이어서 훈련 실시

```python
sc.partial_fit(train_scaled, train_target) # 앞선 학습에 1에포크 이어서 훈련
```

* 훈련 세트와 테스트 세트의 점수 확인

```python
sc.score(train_scaled, train_target)
```

```python
sc.score(test_scaled, test_target)
```

### 에포크와 과대/과소적합
* 에포크 횟수에 따라 과대/과소적합이 발생될 수 있음
* 너무 적은 에포크 횟수 : 과소적합
  - 훈련 세트 학습이 덜 되어 훈련 세트와 테스트 세트에 잘 맞지 않음
* 너무 많은 에포크 횟수 : 과대적합
  - 훈련 세트에 너무 치중하여 학습

* 과대적합이 발생하기 직전에 훈련을 멈추기(조기 종료: Early Stopping) 위한 최적의 에포크(Epoch) 찾기

```python
import numpy as np

sc = SGDClassifier(loss = 'log_loss', random_state =42)
train_score = []
test_score=[]

# partial_fit() 메소드를 fit() 메서드 쓰지 않고 바로 호출하면
# partial_fit() 메서드에 분류할 티깃 클래스가 뭔지 알려줘야함
classes = np.unique(train_target)
```

* 300번의 에포크 동안 훈련을 반복하고 훈련 세트와 테스트 세트의 점수를 계산하여 train_score, test_score 리스트에 추가

```python
for _ in range(0,300):
    sc.partial_fit(train_scaled, train_target, classes = classes)
    train_score.append(sc.score(train_scaled, train_target))
    test_score.append(sc.score(test_scaled, test_target))
```

* 훈련 세트와 테스트 세트의 점수를 그래프로 시각화

```python
import matplotlib.pyplot as plt

plt.plot(train_score, color='red')
plt.plot(test_score, color='blue')
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.show()
```

* 최적의 반복 횟수(에포크)를 선택하여 SGDClassifier 클래스를 이용하여 분류 모델 생성

```python
# tol=None으로 설정: 훈련을 자동으로 조기 종료하지 않고 max_iter만큼 무조건 반복
sc = SGDClassifier(loss = 'log_loss', max_iter=100, tol=None, random_state=42)
sc.fit(train_scaled, train_target)
```

```python
sc.score(train_scaled, train_target)
```

```python
sc.score(test_scaled, test_target)
```
